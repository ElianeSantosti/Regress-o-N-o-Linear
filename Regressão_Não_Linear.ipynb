```python
import pandas as pd
import numpy as np
```

```python
casa = pd.read_excel("/content/house (1).xlsx")
casa.head()
```

```python
casa.info()
```

```python
casa = casa.dropna()
```

```python
casa.describe()
```

```python
#criando  variÃ¡veis dummy
casa = pd.get_dummies(casa)
casa.head()
```

```python
import numpy as np

labels = np.array(casa['SalePrice'])   #criando um array   extraindo os preÃ§os de venda da coluna 'SalePrice' e armazenando-os em um formato NumPy.

features= casa.drop('SalePrice', axis = 1)  #  excluindo a coluna 'SalePrice

feature_list = list(casa.columns)  #criando uma lista de nomes de colunas do DataFrame 'casa' e armazenando-a na variÃ¡vel 'feature_list

casa = np.array(features)    #    convertendo o DataFrame 'features' em um array NumPy e atribuindo-o Ã  variÃ¡vel 'casa'. Isso transforma os recursos (colunas) do DataFrame em uma matriz NumPy,
```

```python
#Usando Skicit-learn para dividir dados em conjuntos de treinamento e teste
from sklearn.model_selection import train_test_split
# Dividindo  os dados em conjuntos de treinamento e teste
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

# Bibliotecas

```python
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostClassifier,GradientBoostingClassifier
from sklearn import metrics
```

# Treino com RandomForestRegressor

construÃ­do a partir de vÃ¡rias Ã¡rvores de decisÃ£o. Cada Ã¡rvore de decisÃ£o Ã© um modelo simples que divide os dados em subgrupos com base nas caracterÃ­sticas dos dados. As Ã¡rvores de decisÃ£o sÃ£o construÃ­das de forma que cada nÃ³ da Ã¡rvore representa uma decisÃ£o ou uma regra

```python

from sklearn.ensemble import RandomForestRegressor
#define o nÃºmero de Ã¡rvores de decisÃ£o
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)   #O random_state Ã© uma semente que controla a aleatoriedade no treinamento das Ã¡rvores e na divisÃ£o dos dados.

rf.fit(train_features, train_labels);   # treinando o modelo
```

```python
# Usando  forest's
predictions_rf = rf.predict(test_features)

# Calculando erros  abasolutos
errors_rf = abs(predictions_rf - test_labels)

r_sq = rf.score(features, labels)
print('Coeficiente de DeterminaÃ§Ã£o (RÂ²):', r_sq)

print('MAE:', metrics.mean_absolute_error(test_labels, predictions_rf))
print('MSE:', metrics.mean_squared_error(test_labels, predictions_rf))
print('RMSE:', np.sqrt(metrics.mean_squared_error(test_labels, predictions_rf)))
```
