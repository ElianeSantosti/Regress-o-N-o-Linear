```python
import pandas as pd
import numpy as np
```

```python
casa = pd.read_excel("/content/house (1).xlsx")
casa.head()
```

```python
casa.info()
```

```python
casa = casa.dropna()
```

```python
casa.describe()
```

```python
#criando  variáveis dummy
casa = pd.get_dummies(casa)
casa.head()
```

```python
import numpy as np

labels = np.array(casa['SalePrice'])   #criando um array   extraindo os preços de venda da coluna 'SalePrice' e armazenando-os em um formato NumPy.

features= casa.drop('SalePrice', axis = 1)  #  excluindo a coluna 'SalePrice

feature_list = list(casa.columns)  #criando uma lista de nomes de colunas do DataFrame 'casa' e armazenando-a na variável 'feature_list

casa = np.array(features)    #    convertendo o DataFrame 'features' em um array NumPy e atribuindo-o à variável 'casa'. Isso transforma os recursos (colunas) do DataFrame em uma matriz NumPy,
```

```python
#Usando Skicit-learn para dividir dados em conjuntos de treinamento e teste
from sklearn.model_selection import train_test_split
# Dividindo  os dados em conjuntos de treinamento e teste
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

# Bibliotecas

```python
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostClassifier,GradientBoostingClassifier
from sklearn import metrics
```

# Treino com RandomForestRegressor

construído a partir de várias árvores de decisão. Cada árvore de decisão é um modelo simples que divide os dados em subgrupos com base nas características dos dados. As árvores de decisão são construídas de forma que cada nó da árvore representa uma decisão ou uma regra

```python

from sklearn.ensemble import RandomForestRegressor
#define o número de árvores de decisão
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)   #O random_state é uma semente que controla a aleatoriedade no treinamento das árvores e na divisão dos dados.

rf.fit(train_features, train_labels);   # treinando o modelo
```

```python
# Usando  forest's
predictions_rf = rf.predict(test_features)

# Calculando erros  abasolutos
errors_rf = abs(predictions_rf - test_labels)

r_sq = rf.score(features, labels)
print('Coeficiente de Determinação (R²):', r_sq)

print('MAE:', metrics.mean_absolute_error(test_labels, predictions_rf))
print('MSE:', metrics.mean_squared_error(test_labels, predictions_rf))
print('RMSE:', np.sqrt(metrics.mean_squared_error(test_labels, predictions_rf)))
```

# Visualização de árvores - Random Forest

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot


tree = rf.estimators_[5]   # # Selecionar a sexta árvore da Floresta Aleatória

# Get the feature names
feature_list = [str(i) for i in range(len(tree.feature_importances_))]  # Cria uma lista de nomes de características como strings

# Exportar a imagem para um arquivo DOT
export_graphviz(tree, out_file='tree.dot', feature_names=feature_list, rounded=True, precision=1) # Exportar a árvore de decisão para um arquivo DOT ('tree.dot') para visualização
# Usar nomes de características para uma melhor interpretação
# 'rounded=True' arredonda os cantos dos nós da árvore de decisão
# 'precision=1' define a precisão dos valores exibidos para uma casa decimal

(graph, ) = pydot.graph_from_dot_file('tree.dot')

# Write the graph to a png file
graph.write_png('tree.png')
```

# Feature importance - Random Forest
